# Algerian Forest Fires Dataset 
## Dataset Information

The dataset contains **244 instances** collected from two regions in Algeria:

- **Bejaia Region** (Northeast Algeria)  
- **Sidi Bel-Abb√®s Region** (Northwest Algeria)

Each region contributes **122 instances**, making a balanced regional distribution.

### Time Period
The data was recorded between **June 2012 and September 2012**.

### Features
The dataset includes:

- **11 input attributes**  
- **1 output attribute (class label)**

### Class Distribution
The 244 instances are classified into:

- **Fire:** 138 instances  
- **Not Fire:** 106 instances  

Overall, the dataset provides environmental and meteorological variables useful for wildfire prediction and analysis.


Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

# DATA CLEANING

### Removing null values

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df.head()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
df.iloc[[122]]
```

```python
##remove the 122nd row
df=df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.isnull().sum()
```

### Changing data type

```python
df.head()
```

```python
df.columns
```

```python
#stripping the extra spaces in columns
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
df[['Region','day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = \
    df[['Region','day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']] = \
    df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
df.info()
```

```python
df.head()
```

```python
## Let ave the cleaned dataset
df.to_csv('data/Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

The data shows that August and September recorded the highest number of forest fires across both regions. From the monthly distribution plot, we can observe the following:

August experienced the largest number of fires, making it the peak month for wildfire occurrences.

Most fires occurred within a three-month window: June, July, and August, indicating a concentrated fire season.

September had comparatively fewer fires, though still higher than other months outside this peak period.
